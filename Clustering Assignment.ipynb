### Theoretical Questions:

1. What is unsupervised learning in the context of machine learning   
    * Unsupervised learning is a type of machine learning where the model is trained on unlabeled data. The algorithm tries to find   patterns, structures, or relationships in the data without predefined outputs. Common tasks in unsupervised learning include clustering, dimensionality reduction, and anomaly detection.

2. What is the difference between supervised and unsupervised learning
    1. Select k initial cluster centroids randomly.
    2. Assign each data point to the nearest centroid based on Euclidean distance.
    3. Compute new centroids by averaging the points in each cluster.
    4. Repeat steps 2 and 3 until centroids no longer change significantly or a maximum number of iterations is reached.

3. Explain the concept of a dendrogram in hierarchical clustering
    * A dendrogram is a tree-like diagram that illustrates the arrangement of clusters formed during hierarchical clustering. It shows how data points are merged or split at different levels of hierarchy, allowing the user to visualize relationships between clusters and determine an appropriate number of clusters.

4. What is the main difference between K-Means and Hierarchical Clustering
    * The key difference is that K-Means is a partitioning algorithm that assigns data points into k clusters, whereas Hierarchical Clustering creates a hierarchy of clusters using agglomerative (bottom-up) or divisive (top-down) approaches. Hierarchical clustering does not require specifying the number of clusters in advance, unlike K-Means.

5. What are the advantages of DBSCAN over K-Means
    * Does not require specifying k clusters in advance.
    * Can find clusters of arbitrary shape, unlike K-Means, which assumes spherical clusters.
    * Handles noise effectively by identifying outliers.
    * Works well with varying densities, whereas K-Means struggles with clusters of different densities.

6. When would you use Silhouette Score in clustering
    * Silhouette Score is used to evaluate the quality of clustering by measuring how well-separated the clusters are. It is useful fo
        * Comparing different clustering algorithms.
        * Choosing the optimal number of clusters in K-Means.
        * Identifying whether clusters are well-defined.

7. What are the limitations of Hierarchical Clustering
    * Computationally expensive: (2)complexity makes it inefficient for large datasets.
    * Not flexible: Once a merge/split occurs, it cannot be undone.
    * Sensitive to noise and outliers, which can lead to poor clustering results.
    * Difficult to scale due to memory and computation constraints.

8. Why is feature scaling important in clustering algorithms like K-Means
    * Feature scaling ensures that all features contribute equally to distance calculations. Since K-Means relies on Euclidean distance, features with larger ranges can dominate the clustering process, leading to biased results. Standardization (z-score normalization) or Min-Max scaling is often applied.

9. How does DBSCAN identify noise points
    * Core points: Have at least M in Pts neighbors within radius 系.
    * Border points: Have fewer than  ○  neighbors but are reachable from a core point.
    * Noise points: Neither core nor border points, meaning they do not belong to any cluster.

10. Define inertia in the context of K-Means
    * Inertia is the sum of squared distances of each data point to its closest cluster centroid. It measures how tightly the data points are grouped within clusters. A lower inertia value indicates better clustering, but decreasing inertia too much may lead to overfitting.
    
11. What is the elbow method in K-Means clustering
    * The elbow method is a technique used to determine the optimal number of clusters (k) in K-Means. It involves plotting inertia (sum of squared distances from points to their closest centroid) against different values of k. The "elbow point" is where the rate of decrease in inertia slows down, indicating the best balance between compact clusters and overfitting.

12. Describe the concept of "density" in DBSCAN
    * In DBSCAN (Density-Based Spatial Clustering of Applications with Noise), density refers to the number of points within a specified radius (系). A region is considered dense if it has at least a minimum number of points (○) within 系. Clusters are formed by connecting dense regions, and sparse regions are treated as noise.

13. Can hierarchical clustering be used on categorical data
    * Yes, hierarchical clustering can be used on categorical data, but it requires an appropriate distance metric
        * Hamming distance (for binary attributes)
        * Jaccard similarity (for categorical attributes)
        * Gower's distance (for mixed data types)Alternatively, categorical data can be encoded using techniques like one-hot encoding before applying standard hierarchical clustering.

14. What does a negative Silhouette Score indicate
    * A negative Silhouette Score means that a data point is assigned to the wrong clusterit is closer to points in a different cluster than its own. This suggests poor clustering and possible overlap between clusters.

15. Explain the term "linkage criteria" in hierarchical clustering
    * Linkage criteria define how distances between clusters are calculated when merging them. Common linkage methods include:
        * Single linkage: Distance between the closest points in two clusters.
        * Complete linkage: Distance between the farthest points in two clusters.
        * Average linkage: Mean distance between all points in two clusters.
        * Centroid linkage: Distance between the centroids of two clusters.
        * Different linkage criteria affect the shape and structure of the resulting clusters.

16. Why might K-Means clustering perform poorly on data with varying cluster sizes or densities
    * Clusters of different densities: Since it minimizes variance, smaller dense clusters may be overshadowed by larger sparse ones.
    * Clusters of different sizes: K-Means treats all clusters equally, so large clusters may distort centroid positions.
    * Non-spherical clusters: K-Means is ineffective for clusters with irregular shapes.

17. What are the core parameters in DBSCAN, and how do they influence clustering
    * (Epsilon): The maximum distance within which points are considered neighbors. Larger 系 values form larger clusters.MinPts: The minimum number of points required to form a dense region. Smaller 
    * MinPts values lead to more clusters, while larger values require denser regions.These parameters influence how DBSCAN detects clusters and differentiates them from noise.

18. How does K-Means++ improve upon standard K-Means initialization
    * K-Means++ enhances standard K-Means by selecting initial centroids more intelligently to avoid poor cluster assignments. Instead of random initialization
        * Chooses the first centroid randomly.
        * Selects subsequent centroids with a probability proportional to their squared distance from existing centroids.
        * This spreads out the initial centroids, leading to faster convergence and better clustering results.

19. What is agglomerative clustering
    * Agglomerative clustering is a bottom-up hierarchical clustering method where each data point starts as its own cluster, and clusters are merged iteratively based on a linkage criterion until a single cluster remains. This approach builds a dendrogram, which can be cut at different levels to obtain desired clusters.

20. What makes Silhouette Score a better metric than just inertia for model evaluation?
    * Inertia measures only compactness but does not consider separation between clusters.
    * Silhouette Score evaluates both cohesion and separation, giving a more comprehensive assessment of clustering quality.
    * A higher Silhouette Score indicates well-defined, well-separated clusters, while inertia may suggest a misleading improvement simply by increasing the number of clusters.

```python
import pandas as pd 
import numpy as np
import seaborn as sns 
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs  , make_moons , load_iris , load_wine , make_circles ,load_breast_cancer , load_digits
from sklearn.cluster import KMeans , AgglomerativeClustering ,  DBSCAN
from sklearn.preprocessing import StandardScaler , MinMaxScaler
from sklearn.decomposition import PCA
from sklearn.metrics import silhouette_samples, silhouette_score
```

```python
# 21. Generate synthetic data with 4 centers using make_blobs and apply K-Means clustering. Visualize using a scatter plot

X_blobs, y = make_blobs(n_samples=300, centers=4, cluster_std=1.0, random_state=1)

kmeans = KMeans(n_clusters=4)
y_kmeans = kmeans.fit_predict(X_blobs)

plt.figure(figsize=(8, 6))
plt.scatter(X_blobs[:, 0], X_blobs[:, 1], c=y_kmeans)

plt.title("K-Means Clustering on Synthetic Data")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.show()
```

```python
# 22. Load the Iris dataset and use Agglomerative Clustering to group the data into 3 clusters. Display the first 10predicted labels

iris = load_iris()
X_iris = iris.data

agg_clustering = AgglomerativeClustering(n_clusters=3)
y_agg = agg_clustering.fit_predict(X_iris)

print(y_agg[:10])
```

```python
# 23. Generate synthetic data using make_moons and apply DBSCAN. Highlight outliers in the plot

X_moons, y_moons = make_moons(n_samples=300, noise=0.1, random_state=1)

dbscan = DBSCAN(eps=0.2, min_samples=5)
y_dbscan = dbscan.fit_predict(X_moons)

plt.figure(figsize=(8, 6))
plt.scatter(X_moons[:, 0], X_moons[:, 1], c=y_dbscan, cmap='viridis', alpha=0.6, edgecolors='k')

outliers = X_moons[y_dbscan == -1]
plt.scatter(outliers[:, 0], outliers[:, 1], c='red', marker='x', label='Outliers', s=100)


plt.title("DBSCAN Clustering on Moons Data")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.show()
```
