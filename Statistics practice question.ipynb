````python

int_list = np.random.randint(90,130 , size = 100  )

# i) Write a Python function to calculate the mean of a given list of numbers. Create a function to find the median of a list of numbers.

mean = np.mean(int_list)
print ("int_list mean = ", mean)

median = np.median(int_list)
print ("int_list median = ", median )

# ii) Develop a program to compute the mode of a list of integers.  

mode = s.mode(int_list)
print("int_list mode = ", mode)

# iii) Implement a function to calculate the weighted mean of a list of values and their corresponding weights. 

def weighted_mean(values, weights):
    # Check if the lengths of the values and weights lists are the same
    if len(values) != len(weights):
        raise ValueError("The number of values and weights must be the same.")
    
    # Calculate the sum of the weighted values and the sum of the weights
    weighted_sum = sum(v * w for v, w in zip(values, weights))
    total_weight = sum(weights)
    
    # Calculate and return the weighted mean
    if total_weight == 0:
        raise ValueError("The total weight cannot be zero.")
    
    return weighted_sum / total_weight

# iv) Write a Python function to find the geometric mean of a list of positive numbers.

def geometric_mean(numbers):
    if len(numbers) == 0:
        return 0  
    
    product = 1
    for number in numbers:
        product *= number
    
    return product ** (1 / len(numbers))
 
# v) Create a program to calculate the harmonic mean of a list of values

def harmonic_mean(values):
   
    if len(values) == 0:
        raise ValueError("The list of values cannot be empty")
    
    # Calculate the sum of reciprocals of the values
    reciprocal_sum = sum(1 / v for v in values if v != 0)  # Skip division by zero
    
    # Ensure no zero values were present, as harmonic mean is undefined for zero
    if reciprocal_sum == 0:
        raise ValueError("Harmonic mean is undefined for lists with zero values")
    
    # Calculate the harmonic mean
    n = len(values)
    return n / reciprocal_sum

# vi) Build a function to determine the midrange of a list of numbers (average of the minimum and maximum).

def midrange(values):
   
    if len(values) == 0:
        raise ValueError("The list of values cannot be empty")
    
    # Calculate the minimum and maximum values in the list
    minimum = min(values)
    maximum = max(values)
    
    # Calculate the midrange
    return (minimum + maximum) / 2

# vii) Implement a Python program to find the trimmed mean of a list, excluding a certain percentage of outliers. 

def trimmed_mean(values, trim_percentage):
    
    if len(values) == 0:
        raise ValueError("The list of values cannot be empty")
    
    if not (0 <= trim_percentage <= 50):
        raise ValueError("Trim percentage must be between 0 and 50")
    
    # Sort the values in ascending order
    values_sorted = sorted(values)
    
    # Calculate how many values to trim
    n_trim = int(len(values) * trim_percentage / 100)
    
    # Remove the smallest and largest values (outliers)
    trimmed_values = values_sorted[n_trim:len(values) - n_trim]
    
    # Calculate the mean of the remaining values
    return sum(trimmed_values) / len(trimmed_values)
````

3.  Write a Python class representing a discrete random variable with methods to calculate its expected value and variance.

````python
class DiscreteRandomVariable:
    def __init__(self, values, probabilities):
        
        self.values = values
        self.probabilities = probabilities

    def expected_value(self):
        return sum(v * p for v, p in zip(self.values, self.probabilities))

    def variance(self):

        expected_val = self.expected_value()
        expected_val_squared = sum((v ** 2) * p for v, p in zip(self.values, self.probabilities))
        return expected_val_squared - expected_val ** 2
````

4. Implement a program to simulate the rolling of a fair six-sided die and calculate the expected value and
variance of the outcomes.

````python
import random

def roll_die():
    return random.randint(1, 6)

# Calculate the theoretical expected value (mean) of the die outcomes
def expected_value():
    outcomes = [1, 2, 3, 4, 5, 6]
    probabilities = [1/6] * 6  # Fair die, each outcome has probability 1/6
    return sum(x * p for x, p in zip(outcomes, probabilities))

# Calculate the theoretical variance of the die outcomes
def variance():
    outcomes = [1, 2, 3, 4, 5, 6]
    probabilities = [1/6] * 6  # Fair die, each outcome has probability 1/6
    
    # Calculate E(X^2)
    expected_squared = sum((x**2) * p for x, p in zip(outcomes, probabilities))
    
    # Calculate E(X)
    mean = expected_value()
    
    # Variance is E(X^2) - (E(X))^2
    return expected_squared - mean**2

# Simulate rolling the die multiple times and compute empirical expected value and variance
def simulate_rolls(num_rolls):
    rolls = [roll_die() for _ in range(num_rolls)]
    mean = sum(rolls) / num_rolls
    squared_deviations = [(x - mean) ** 2 for x in rolls]
    var = sum(squared_deviations) / num_rolls
    return mean, var

# Main function to display results
def main():
    # Theoretical expected value and variance
    theoretical_mean = expected_value()
    theoretical_variance = variance()
    
    print(f"Theoretical Expected Value (E(X)): {theoretical_mean}")
    print(f"Theoretical Variance (Var(X)): {theoretical_variance}")
    
    # Simulate rolling the die 10,000 times
    num_rolls = 10000
    simulated_mean, simulated_variance = simulate_rolls(num_rolls)
    
    print(f"Simulated Expected Value (E(X)) with {num_rolls} rolls: {simulated_mean}")
    print(f"Simulated Variance (Var(X)) with {num_rolls} rolls: {simulated_variance}")

# Run the program
main()
````

5. Create a Python function to generate random samples from a given probability distribution (e.g.,
binomial, Poisson) and calculate their mean and variance.

````python
def generate_random_samples(distribution_type, params, size):
    
    if distribution_type == 'binomial':
       
        n, p = params
        samples = np.random.binomial(n, p, size)
    
    elif distribution_type == 'poisson':
        lambda_param = params
        samples = np.random.poisson(lambda_param, size)
    
    else:
        raise ValueError("Unsupported distribution type. Choose 'binomial' or 'poisson'.")
    

    mean = np.mean(samples)
    variance = np.var(samples)
    
    return mean, variance

# Example usage:
def main():
    # Parameters for the distributions
    binomial_params = (10, 0.5)  # n=10 trials, p=0.5 probability of success
    poisson_params = 5  # lambda=5 (mean rate of events)
    
    # Size of the samples
    size = 1000
    
    # Generate random samples from Binomial distribution
    binomial_mean, binomial_variance = generate_random_samples('binomial', binomial_params, size)
    print(f"Binomial Distribution (n=10, p=0.5):")
    print(f"  Simulated Mean: {binomial_mean}")
    print(f"  Simulated Variance: {binomial_variance}")
    
    # Generate random samples from Poisson distribution
    poisson_mean, poisson_variance = generate_random_samples('poisson', poisson_params, size)
    print(f"\nPoisson Distribution (lambda=5):")
    print(f"  Simulated Mean: {poisson_mean}")
    print(f"  Simulated Variance: {poisson_variance}")

# Run the example usage
main()
````

````python
def generate_gaussian_samples(mean, std_dev, size):

    samples = np.random.normal(mean, std_dev, size)
    return samples

def calculate_statistics(samples):
    
    mean = np.mean(samples)
    variance = np.var(samples)
    std_dev = np.std(samples)
    
    return mean, variance, std_dev

def main():
    # Parameters for the Gaussian distribution
    mean = 0      # Mean (μ) of the distribution
    std_dev = 1   # Standard deviation (σ) of the distribution
    size = 1000   # Number of samples to generate
    
    # Generate random samples from the Gaussian distribution
    samples = generate_gaussian_samples(mean, std_dev, size)
    
    # Calculate the mean, variance, and standard deviation of the samples
    mean, variance, std_dev = calculate_statistics(samples)
    
    # Output the results
    print(f"Generated {size} samples from a Gaussian distribution with mean={mean} and standard deviation={std_dev}")
    print(f"Sample Mean: {mean}")
    print(f"Sample Variance: {variance}")
    print(f"Sample Standard Deviation: {std_dev}")

# Run the program
main()
````

````python
df = sns.load_dataset("tips")
# df.info()
# (i) Write a Python function that calculates their skewness.    

def calculate_skewness(data ):
    n = len(data)
    mean = np.mean(data)
    std_dev = np.std(data, ddof=0)  # Population standard deviation

    skewness = (n / ((n-1)*(n-2))) * np.sum(((data - mean) / std_dev) ** 3)
    
    return skewness

skewness = calculate_skewness(df["total_bill"] )
print(f"skewness of total_bill{skewness}")

# (ii) Create a program that determines whether the columns exhibit positive skewness, negative skewness, or is approximately symmetric. 

from scipy.stats import skew

def skewness_type(skew_value):
    if skew_value > 0:
        return 'Positive Skewness (Right-skewed)'
    elif skew_value < 0:
        return 'Negative Skewness (Left-skewed)'
    else:
        return 'Symmetric (Approximately Normal)'

# Calculate skewness for each column and determine its type
for column in df.columns:
    skew_value = skew(df["total_bill"])

    print(f"{column}: Skewness = {skew_value:.2f} => {skewness_type(skew_value)}")


# iii)  Write a function that calculates the covariance between two columns. 

def calculate_covariance(X, Y):
    # Ensure the inputs are numpy arrays for easy manipulation
    X = np.array(X)
    Y = np.array(Y)
    
    # Calculate the means of X and Y
    mean_X = np.mean(X)
    mean_Y = np.mean(Y)
    
    # Calculate covariance using the formula
    covariance = np.mean((X - mean_X) * (Y - mean_Y))
    
    return covariance

covariance = calculate_covariance(df["total_bill"],df["tip"])

print(f"covariance of toall_bill and tip = { covariance}")

# iv)  Implement a Python program that calculates the Pearson correlation coefficient between two columns.  

def pearson_correlation_numpy(X, Y):
    # Using numpy's corrcoef function, which returns a correlation matrix
    correlation_matrix = np.corrcoef(X, Y)
    
    # The correlation coefficient is located at position [0, 1] in the correlation matrix
    return correlation_matrix[0, 1]

correlation_numpy = pearson_correlation_numpy(df["total_bill"],df["tip"])
print(f"Pearson correlation coefficient toall_bill and tip =  {correlation_numpy:.2f}")

# v) Write a script to visualize the correlation between two specific columns in a Pandas DataFrame using scatter plots.  

def visualize_correlation(df, col1, col2):
    # Set the style for seaborn
    sns.set(style="whitegrid")
    
    # Create a scatter plot with seaborn
    plt.figure(figsize=(8, 6))  
    sns.scatterplot(x=col1, y=col2, data=df, color='blue', s=100) 
    
    # Optionally, add a regression line to visualize linear correlation
    sns.regplot(x=col1, y=col2, data=df, scatter=False, color='red', line_kws={'lw': 2, 'ls': '--'})
    
    # Set plot title and labels
    plt.title(f"Scatter Plot: {col1} vs {col2}", fontsize=14)
    plt.xlabel(col1, fontsize=12)
    plt.ylabel(col2, fontsize=12)
    
    # Show plot
    plt.show()


# Call the function to visualize the correlation between 'Column1' and 'Column2'
visualize_correlation(df, "total_bill","tip")
````

8. Write a Python function to calculate the probability density function (PDF) of a continuous random
variable for a given normal distribution.

````python
def normal_pdf(x, mean, std_dev):

    # Compute the PDF using the formula for a normal distribution
    pdf_value = (1 / (std_dev * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mean) / std_dev) ** 2)
    return pdf_value

# Example usage
def main():
    # Parameters for the normal distribution
    mean = 0      # Mean (μ)
    std_dev = 1   # Standard deviation (σ)
    
    # Points to evaluate the PDF at
    x_values = [-2, -1, 0, 1, 2]
    
    # Calculate the PDF for each x value
    for x in x_values:
        pdf_value = normal_pdf(x, mean, std_dev)
        print(f"PDF at x={x}: {pdf_value}")

# Run the program
main()
````

9. Create a program to calculate the cumulative distribution function (CDF) of exponential distribution.

````python


def exponential_cdf(x, rate_param):

    # Compute the CDF using the formula for an exponential distribution
    cdf_value = 1 - np.exp(-rate_param * x)
    return cdf_value

# Example usage
def main():
    # Parameters for the exponential distribution
    rate_param = 1.0  # Rate parameter λ (mean = 1 / λ)
    
    # Points to evaluate the CDF at
    x_values = [0, 1, 2, 3, 5]
    
    # Calculate the CDF for each x value
    for x in x_values:
        cdf_value = exponential_cdf(x, rate_param)
        print(f"CDF at x={x}: {cdf_value}")

# Run the program
main()
````

10. Write a Python function to calculate the probability mass function (PMF) of Poisson distribution

````python
import math

def poisson_pmf(k, lambd):
   
    # Ensure k is a non-negative integer
    if k < 0 or not isinstance(k, int):
        raise ValueError("k must be a non-negative integer")
    
    # Calculate the Poisson PMF using the formula
    pmf = (lambd**k * math.exp(-lambd)) / math.factorial(k)
    return pmf

# Example usage:
k = 3  # Number of events
lambda_ = 2.5  # Mean rate of events (lambda)

# Calculate PMF for Poisson distribution
probability = poisson_pmf(k, lambda_)
print(f"Poisson PMF for k={k} and lambda={lambda_}: {probability:.4f}")
````

```python
after_program = np.array([80, 85, 90, 80, 92, 80, 95, 90, 85, 88])

```

Use z-test to find if the claims made by tutor are true or false.

````python
from scipy import stats

# Scores before and after the program
before_program = np.array([75, 80, 85, 70, 90, 78, 92, 88, 82, 87])
after_program = np.array([80, 85, 90, 80, 92, 80, 95, 90, 85, 88])

# Calculate differences
differences = after_program - before_program

# Calculate the mean difference
mean_diff = np.mean(differences)

# Assumed population standard deviation of the differences
sigma = 5  # Assumed population standard deviation (replace this with the actual value if known)

# Number of students
n = len(differences)

# Standard Error of the Mean (SEM)
sem = sigma / np.sqrt(n)

# Z-statistic
z_statistic = mean_diff / sem

# Output results
print("Mean difference:", mean_diff)
print("Standard Error of the Mean (SEM):", sem)
print("Z-statistic:", z_statistic)

# Critical value for 95% confidence (two-tailed test)
alpha = 0.05
z_critical = stats.norm.ppf(1 - alpha/2)

# Compare z-statistic with critical value
if abs(z_statistic) > z_critical:
    print("Reject the null hypothesis: The tutoring program has improved the exam scores.")
else:
    print("Fail to reject the null hypothesis: No significant improvement in exam scores.")
````

13) A pharmaceutical company wants to determine if a new drug is effective in reducing blood pressure. They
conduct a study and record blood pressure measurements before and after administering the drug.

```python

after_drug = np.array([130, 140, 132, 128, 145, 148, 138, 136, 125, 130])

  ```


Implement z-test to find if the drug really works or not.

````python

# Blood pressure measurements before and after the drug
before_drug = np.array([145, 150, 140, 135, 155, 160, 152, 148, 130, 138])
after_drug = np.array([130, 140, 132, 128, 145, 148, 138, 136, 125, 130])

# Calculate the differences (before - after)
differences = before_drug - after_drug

# Calculate the mean of the differences
mean_diff = np.mean(differences)

# Calculate the standard deviation of the differences
std_diff = np.std(differences, ddof=1)

# Number of samples
n = len(differences)

# Standard Error of the Mean (SEM)
sem = std_diff / np.sqrt(n)

# Calculate the t-statistic
t_statistic = mean_diff / sem

# Degrees of freedom
df = n - 1

# Calculate the p-value for a one-tailed test (assuming the drug is expected to reduce blood pressure)
p_value = stats.t.cdf(t_statistic, df)

# Output results
print("Mean Difference:", mean_diff)
print("Standard Deviation of Differences:", std_diff)
print("Standard Error of the Mean (SEM):", sem)
print("T-statistic:", t_statistic)
print("Degrees of Freedom:", df)
print("P-value:", p_value)

# Critical value for 95% confidence (one-tailed)
alpha = 0.05
t_critical = stats.t.ppf(1 - alpha, df)

# Decision rule
if t_statistic > t_critical:
    print("Reject the null hypothesis: The drug is effective in reducing blood pressure.")
else:
    print("Fail to reject the null hypothesis: No significant reduction in blood pressure.")

````

````python
def ab_test_analysis(group_A, group_B):
    # Sample sizes
    n_A = len(group_A)
    n_B = len(group_B)
    
    # Sample means
    mean_A = np.mean(group_A)
    mean_B = np.mean(group_B)
    
    # Sample standard deviations
    std_A = np.std(group_A, ddof=1)
    std_B = np.std(group_B, ddof=1)
    
    # Calculate the t-statistic
    t_statistic = (mean_A - mean_B) / np.sqrt((std_A**2 / n_A) + (std_B**2 / n_B))
    
    # Calculate the degrees of freedom
    df = ((std_A**2 / n_A + std_B**2 / n_B)**2) / ((std_A**2 / n_A)**2 / (n_A - 1) + (std_B**2 / n_B)**2 / (n_B - 1))
    
    # Calculate the p-value (two-tailed test)
    p_value = 2 * (1 - stats.t.cdf(abs(t_statistic), df))  # Two-tailed test
    
    # Print results
    print(f"Mean of group A: {mean_A}")
    print(f"Mean of group B: {mean_B}")
    print(f"T-statistic: {t_statistic}")
    print(f"Degrees of freedom: {df}")
    print(f"P-value: {p_value}")
    
    # Decision based on p-value
    alpha = 0.05  # Significance level
    if p_value < alpha:
        print("Reject the null hypothesis: There is a significant difference between the two layouts.")
    else:
        print("Fail to reject the null hypothesis: There is no significant difference between the two layouts.")



layout_a_clicks = np.array([28, 32, 33, 29, 31, 34, 30, 35, 36, 37])  
layout_b_clicks = np.array([40, 41, 38, 42, 39, 44, 43, 41, 45, 47])  


ab_test_analysis(layout_a_clicks, layout_b_clicks)
````
